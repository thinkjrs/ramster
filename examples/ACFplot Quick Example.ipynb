## ACFplot.py Quick Example 

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
# make the plots better within the notebook
%matplotlib inline
```

```python
qqq = pd.read_csv('qqq_2014-12-31.csv')
qqq['Log Close'] = np.log(qqq['Close'])
xt = pd.Series(np.diff(qqq['Log Close']))
xt.name = '1st Differences Log QQQ Close Price'
```

```python
# just grab the function here
# make sure you have the two scripts in this workbook's directory: 'ACFplot.py' and 'ACFplotExceptions.py'
from ACFplot import ACFplot # the script will call the exception classes
"""
EXAMPLE USAGE
"""
# Basic usage
ACFplot(xt)
# Notice how that produces a shitty title. Add a title, custom z values and specify lags for more precision.
ACFplot(np.exp(abs(qqq['Open'] - qqq['Close'])), maxlags=12, title="Open less Closing Price of QQQ", z1 = 1.96, z2 = 1.645)
# Now we'll make it throw an exception by putting in a negative lag length
ACFplot(xt, maxlags=-1)
# As you can see, an exception will kill the output and spit info to the console
# Now let's throw another one...
ACFplot(xt, maxlags=1000000)
```

####  ACF Function
Lag-$l$ Sample ACF Function
$$\hat{\rho} = \frac{ \sum_{t = l + 1}^T(r_t - \bar{r})(r_{t - l} - \bar{r})}{\sum_{t = l}^T(r_t - \bar{r})^2}$$ 


\*\*Tsay, Ruey S. - Analysis of Financial Time Series, 3rd ed.
